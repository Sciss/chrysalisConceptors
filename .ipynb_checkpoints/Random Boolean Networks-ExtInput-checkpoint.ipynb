```python
from functools import partial
def bytesToInt(b):
    return reduce(lambda a,x: (a<<1) + int(x), b)

class RBN():
    @staticmethod
    def create(N,K,I=0):
        state = {}
        state['N'] = N
        state['K'] = K
        state['I'] = I 
        state['nodes'] = np.random.rand(N) > 0.5 
        state['cx'] = [np.random.choice(N,K, replace=False) for x in zeros(N)]
        if (I > 0):
            state['inodes'] = np.random.rand(I) > 0.5 #zeros(N, dtype=np.bool_)
            state['icx'] = [np.random.choice(I,1, replace=False) for x in zeros(N)]

        return state
    
    @staticmethod
    def createTruthTableCoupling(state, prob):
        tableWidth = state['K'] + (1 if state['I'] > 0 else 0)
        state['funcTables'] = map(lambda x: np.random.rand(pow(2,tableWidth)) > prob, np.zeros(state['N']))
        def truthTableResult(state, nodeIdx, cxVals):
            cxFuncIndex = bytesToInt(cxVals)
            funcResult = state['funcTables'][nodeIdx][cxFuncIndex]
            return funcResult
            
        state['couplingFunction'] = truthTableResult
        return state
    
    @staticmethod
    def createAndCoupling(state):
        def couplingAnd(state, nodeIdx, cxVals):
            return cxVals.sum() == cxVals.size 
        state['couplingFunction'] = couplingAnd
        return state

    @staticmethod
    def createAnyOrCoupling(state):
        def couplingAnyOr(state, nodeIdx, cxVals):
            return cxVals.sum() > 0 
        state['couplingFunction'] = couplingAnyOr
        return state

    @staticmethod
    def createMinCountCoupling(state, maxMinimum):
        def couplingMinCount(state, nodeIdx, cxVals):
            return cxVals.sum() >= state['nodeMinCounts'][nodeIdx]
        
        state['nodeMinCounts'] =[np.random.choice(maxMinimum, 1)[0]+1 for x in zeros(state['N'])]
        state['couplingFunction'] = couplingMinCount
        return state

    
    @staticmethod
    def calculateNode(state, nodeIdx):
        cxVals = state['nodes'][state['cx'][nodeIdx]]
        if (state['I'] > 0):
            ival = state['inodes'][state['icx'][nodeIdx]]
            cxVals = np.append(cxVals, ival)
        funcResult = state['couplingFunction'](state, nodeIdx, cxVals)
        return funcResult

    @staticmethod
    def step(state):
        state['nodes'] = np.array([RBN.calculateNode(state,x) for x in np.arange(state['N'])], dtype=np.bool_)      
        return state

    @staticmethod
    def stepAsync(state):
        seq = np.random.choice(state['N'], state['N'], replace=False)
        return RBN.stepAsyncSeq(state, seq)
    
    @staticmethod
    def stepAsyncSeq(state, seq):
        for i in seq:
            state['nodes'][i] = RBN.calculateNode(state, i)
        return state

    
runlen = 100
state = RBN.createTruthTableCoupling(RBN.create(10,3,0), 0.5)
#state = RBN.createAnyOrCoupling(RBN.create(64,2))
#state = RBN.createMinCountCoupling(RBN.create(128,30),20)
#print state
state['nodes'] = np.random.rand(state['N']) > 0.5
startState=state['nodes']

state['inodes'] = np.random.rand(state['I']) > 0.5
print "running:"
#output = [RBN.stepAsync(state)['nodes'] for x in zeros(runlen)]
output = np.zeros([runlen, state['N']], dtype=np.bool_)
for x in xrange(runlen):
    output[x] = RBN.step(state)['nodes'] 
    
figure(1)
figsize(10,5)
imshow(np.transpose(output), cmap='hot')

figure(2)
plot([sum(x) for x in output])
plot([hammingDistance(x,y) for x,y in zip(output[:-1], output[1:])])


# figure(3)
# state['nodes'] = startState
# state['inodes'] = np.random.rand(state['I']) > 0.5
# output = [RBN.step(state)['nodes'] for x in zeros(runlen)]
# imshow(np.transpose(output), cmap='hot')

# figure(4)
# plot([sum(x) for x in output])
# plot([hammingDistance(x,y) for x,y in zip(output[:-1], output[1:])])

# figure(5)
# state['nodes'] = np.random.rand(state['N']) > 0.5
# output = [RBN.step(state)['nodes'] for x in zeros(runlen)]
# imshow(np.transpose(output), cmap='hot')

# figure(6)
# plot([sum(x) for x in output])
# plot([hammingDistance(x,y) for x,y in zip(output[:-1], output[1:])])
```

```python
def matchEndBlock(v, size):
    result = len(v) >= (size * 2)
    if (result):
        block1 = v[- size:]
        block2 = v[size * -2:- size]
        result = np.array_equal(block1, block2)
    return result

def detectAttractor(v):
    found=False
    blockSize=0
    while (found == False) and (blockSize * 2 < len(v)):
        blockSize=blockSize+1
        found = matchEndBlock(v,blockSize)
    return (found, blockSize)

q = np.array([[0,0],[0,1],[0,1]])
print(q)
detectAttractor(q)
```

```python
def randSwapPair(arr):
    idxs = np.random.choice(len(arr), 2, replace=False)
    tmp = arr[idxs[0]]
    arr[idxs[0]] = arr[idxs[1]]
    arr[idxs[1]] = tmp
    return arr
    
x1  =np.arange(10)
x2 = randSwapPair(x1)
print x1
print x2
```
